```python
from bi import bi
import matplotlib.pyplot as plt
%matplotlib inline
```

Basic graph of Block length over time (up to block 140000).

```python
result = bi("--block %bl --end=140000")

plt.figure(figsize=(20,8))
plt.plot(result["results"]["%bl"], label="%bl")
plt.show()
```

Example of a smaller chunk of block lengths.

```python
result = bi("--block %bl --start=100000 --end=101000")

plt.figure(figsize=(20,8))
plt.plot(result["results"]["%bl"], label="%bl")
plt.show()
```
